# Overview

This is a basic exercise to try and implement VGG-16 using the Keras API, going by the Simonyan ICLR2015 [paper](https://arxiv.org/pdf/1409.1556.pdf).

Uses the tensorflow tutorial from [here](https://www.tensorflow.org/tutorials/quickstart/beginner) as a guide, using a subset of the [Dogs vs Cats](https://www.kaggle.com/c/dogs-vs-cats/data) dataset from kaggle.

## Import data

```python
import tensorflow as tf

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten,\
    Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

import os
import numpy as np
import matplotlib.pyplot as plt


# get the directories of the training images
data_dir = 'C:/users/leahy/Google Drive/Freelance/own_projects/vgg_16_catvdog/'\
          'data/filtered_dataset'

train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'validation')

train_cats_dir = os.path.join(train_dir, 'cats')
train_dogs_dir = os.path.join(train_dir, 'dogs')
val_cats_dir = os.path.join(val_dir, 'cats')
val_dogs_dir = os.path.join(val_dir, 'dogs')

train_cats_num = len(os.listdir(train_cats_dir))
train_dogs_num = len(os.listdir(train_dogs_dir))

val_cats_num = len(os.listdir(val_cats_dir))
val_dogs_num = len(os.listdir(val_dogs_dir))

total_train = train_cats_num + train_dogs_num
total_val = val_cats_num + val_dogs_num
```

```python
print('total training cat images:', train_cats_num)
print('total training dog images:', train_dogs_num)

print('total validation cat images:', val_cats_num)
print('total validation dog images:', val_dogs_num)
print("--")
print("Total training images:", total_train)
print("Total validation images:", total_val)
```

# Preprocessing

We use the `flow_from_directory` method to generate labelled batches of images. The labels are taken from the folder names, and each is a generator which fetches batches of `(X_train, y_train)` pairs from the training folder.

To preview an image, we can call `next()` on the generator to load up a batch from the folder.

```python
batch_size = 128
img_size = 150

train_image_generator = ImageDataGenerator(rescale=1./255)#,
#                                            validation_split=0.9)
val_data_generator = ImageDataGenerator(rescale=1./255)#,
#                                         validation_split=0.9)

train_data_gen = train_image_generator.flow_from_directory(
    batch_size=batch_size, directory=train_dir, shuffle=True,
    target_size=(img_size, img_size), class_mode='binary')#, subset='training')

val_data_gen = train_image_generator.flow_from_directory(
    batch_size=batch_size, directory=val_dir, shuffle=True,
    target_size=(img_size, img_size), class_mode='binary')#, subset='training')
```

```python
first_batch_images, labels = next(train_data_gen)

def plot_images(images_arr, img_labels):
    fig, axes = plt.subplots(1, 5, figsize=(25,25))
    axes = axes.flatten()
    for img, ax, l in zip(images_arr, axes, img_labels):
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(l, fontdict={'fontsize': 30})
    plt.tight_layout()
    plt.show()
    
    
plot_images(first_batch_images[:5], labels[0:5])
```

## Creating the model
Here we implement the VGG-16 model

```python
l2_lambda = 5e-4

model = Sequential([
    Conv2D(64, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda),
           input_shape=(img_size, img_size, 3)),
#     Conv2D(64, (3, 3), padding='same', activation='relu',
#            kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(128, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
#     Conv2D(128, (3, 3), padding='same', activation='relu',
#            kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(256, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
#     Conv2D(256, (3, 3), padding='same', activation='relu',
#            kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
#     Conv2D(256, (3, 3), padding='same', activation='relu',
#            kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(512, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
#     Conv2D(512, (3, 3), padding='same', activation='relu',
#            kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
#     Conv2D(512, (3, 3), padding='same', activation='relu',
#            kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(512, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    Conv2D(512, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    Conv2D(512, (3, 3), padding='same', activation='relu',
           kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    MaxPooling2D(pool_size=(2,2), strides=(2,2)),
    Flatten(),
    Dropout(0.5),
    Dense(1024, activation='relu',
          kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    Dropout(0.5),
    Dense(1024, activation='relu',
          kernel_regularizer=tf.keras.regularizers.l2(l2_lambda)),
    Dense(1, activation='sigmoid')
])

alpha = 10e-4
adam = Adam(lr=alpha)

model.compile(
    optimizer=adam,
    loss=tf.keras.losses.binary_crossentropy,
    metrics=['accuracy'])

model.summary()
```

## Training the Model

Here we include early stopping, as well as including a checkpoint, where the model is saved to disk if the performance on this epoch is greater than the last.

```python
model.losses
```

```python
num_epochs = 2

checkpoint = ModelCheckpoint(
    "vgg16_1.h5", monitor='val_acc', verbose=1, save_weights_only=False,
    mode='auto', save_freq=1)

early = EarlyStopping(monitor='val_acc', min_delta=0, patience=20, verbose=1,
                      mode='auto')

history = model.fit(
    train_data_gen,
    steps_per_epoch=total_train // batch_size,
    epochs=num_epochs,
    validation_data=val_data_gen,
    validation_steps=total_val // batch_size
#     callbacks=[checkpoint,early]
)
```

```python
l2_lambda = 0
base_model = Sequential([
    Conv2D(32, 3, padding='same', activation='relu',
           kernel_initializer='he_uniform',
           input_shape=(img_size,img_size,3)
           ),
    MaxPooling2D(),
    Conv2D(64, 3, padding='same', activation='relu',
           kernel_initializer='he_uniform'),
    MaxPooling2D(),
    Conv2D(128, 3, padding='same', activation='relu',
           kernel_initializer='he_uniform'),
    MaxPooling2D(),
    Flatten(),
    # Dropout(0.3),
    Dense(128, activation='relu',
           kernel_initializer='he_uniform'),
    Dense(1, activation='sigmoid')
])
def compile_model(model):
    opt = tf.keras.optimizers.SGD(momentum=0.9, lr=0.001)

    model.compile(
        optimizer=opt,
        loss='binary_crossentropy',
        metrics=['accuracy'])

    model.summary()

compile_model(base_model)
```
